```python


# Load the image
image_path = 'project.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)

# Convert the image to the HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds for the red color
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 255, 255])

# Create a mask to isolate the red regions in the image
mask = cv2.inRange(hsv_image, lower_red, upper_red)

# Find contours in the mask
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw the detected red balls on the original image
for contour in contours:
    approx = cv2.approxPolyDP(contour, 0.02 * cv2.arcLength(contour, True), True)
    cv2.drawContours(image, [approx], 0, (0, 0, 255), 5)

# Display the original image with detected red balls
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Code Explanation

## Load the Image:

It begins by specifying the path to an image file, which should be replaced with the actual path to your image.
It then uses the OpenCV (cv2) library to read the image and stores it in the variable `image`.

## Convert to HSV Color Space:

The code converts the loaded image from the standard BGR (Blue-Green-Red) color space to the HSV (Hue-Saturation-Value) color space.
HSV is useful for color-based image processing.

## Define the Red Color Range:

It defines a range of red colors in HSV format. The `lower_red` and `upper_red` arrays specify the lower and upper bounds for the red color.
These values are set to detect a range of red hues.

## Create a Mask for Red Regions:

Using the defined color range, the code creates a mask. The mask is a binary image where white pixels represent the regions that match the specified red color range, and black pixels represent all other regions.

## Find Contours:

The code uses the `cv2.findContours` function to identify contours (boundary shapes) within the mask. Contours typically represent connected regions of white pixels.

## Draw Detected Red Regions:

For each detected contour (possibly a red region or red ball), the code approximates its shape using the `cv2.approxPolyDP` function. This helps reduce the number of points that describe the contour while preserving its essential features.
The code then draws the approximated contour on the original image using a red color.

## Display the Result:

Finally, the code displays the original image with the detected red regions highlighted in red. It uses the `matplotlib.pyplot` library (`plt`) to show the image without axis labels.

This code is useful for identifying and visualizing regions in an image that match the specified red color range, potentially highlighting red balls or objects.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = 'project.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)

# Convert the image to the HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds for dark red color
lower_dark_red = np.array([0, 50, 50])  # Adjust these values as needed
upper_dark_red = np.array([5, 255, 255])  # Adjust these values as needed

# Create a mask to isolate the dark red regions in the image
mask = cv2.inRange(hsv_image, lower_dark_red, upper_dark_red)

# Find contours in the mask
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a mask for the detected dark red objects
dark_red_objects_mask = np.zeros_like(image)

# Draw the detected dark red objects on the mask
for contour in contours:
    area = cv2.contourArea(contour)
    if area > 100:  # Filter out small areas to reduce false positives
        cv2.drawContours(dark_red_objects_mask, [contour], 0, (0, 0, 255), -1)

# Bitwise AND the original image with the dark red objects mask
result_image = cv2.bitwise_and(image, dark_red_objects_mask)

# Display the result
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Dark Red Object Detection using OpenCV

This code is designed to detect and highlight dark red objects in an image. It follows these steps:

1. **Import necessary libraries:**
   - `cv2`: OpenCV library for computer vision tasks.
   - `numpy` (aliased as `np`): Used for numerical operations.
   - `matplotlib.pyplot` (aliased as `plt`): Used for displaying the result.

2. **Load the image:**
   - The code specifies the path to an image file (in this case, 'project.jpg') and loads it using OpenCV's `cv2.imread()` function. You should replace 'project.jpg' with the path to your own image.

3. **Convert the image to the HSV color space:**
   - The loaded image is converted from the standard BGR (Blue-Green-Red) color space to the HSV (Hue-Saturation-Value) color space using `cv2.cvtColor()`. HSV is often used for color-based image processing because it separates the color information (hue) from the brightness (value) and saturation.

4. **Define the lower and upper bounds for dark red color:**
   - The code sets a range of HSV values that correspond to the dark red color you want to detect. You can adjust these values to match the specific shade of dark red you are interested in.

5. **Create a mask to isolate dark red regions:**
   - A mask is created using `cv2.inRange()` to filter out pixels in the HSV image that fall within the specified lower and upper bounds for dark red. The result is a binary image where white pixels represent the detected dark red regions, and black pixels represent all other regions.

6. **Find contours in the mask:**
   - `cv2.findContours()` is used to identify contours (boundary shapes) in the binary mask. Contours are typically used to represent connected regions of white pixels.

7. **Create a mask for the detected dark red objects:**
   - A new image called `dark_red_objects_mask` is created with the same dimensions as the original image, initially filled with zeros.

8. **Draw the detected dark red objects on the mask:**
   - The code iterates through the detected contours and checks the area of each contour. Contours with an area greater than a threshold (in this case, 100 pixels) are considered as dark red objects. These objects are then drawn as red-filled shapes on the `dark_red_objects_mask`.

9. **Bitwise AND the original image with the dark red objects mask:**
   - The original image is combined with the `dark_red_objects_mask` using a bitwise AND operation (`cv2.bitwise_and()`). This operation overlays the detected dark red objects onto the original image while keeping other areas black.

10. **Display the result:**
    - The resulting image, which highlights dark red objects in red on the original image, is displayed using `matplotlib.pyplot`.

The code can be customized by adjusting the HSV color range and the minimum area threshold to suit your specific dark red object detection needs.


```python

import cv2
import numpy as np

#Import the image
source = cv2.imread('project.jpg')

#Remove noise from the source image
median = cv2.medianBlur(source, 15)

# Convert BGR to HSV
hsv_source = cv2.cvtColor(median, cv2.COLOR_BGR2HSV)

# Lower red mask
lower_red = np.array([0,100,100])
upper_red = np.array([10,255,255])
red_mask_1 = cv2.inRange(hsv_source, lower_red, upper_red)

# Upper red mask
lower_red = np.array([160,100,100])
upper_red = np.array([180,255,255])
red_mask_2 = cv2.inRange(hsv_source, lower_red, upper_red)

# Join the masks
red_mask_final = red_mask_1 + red_mask_2

# Bitwise-AND mask on the hsv_source image
output_hsv = cv2.bitwise_and(hsv_source, hsv_source, mask = red_mask_final)

# Blur the image to smoothen out bright areas
blurred = cv2.GaussianBlur(output_hsv, (15, 15), 0)

# Threshold the image to reveal bright areas
thresh = cv2.threshold(blurred, 60, 255, cv2.THRESH_BINARY)[1]

# Split the image into components
h_chan, s_chan, v_chan = cv2.split(thresh)

# Erode and dilate the image to remove small unwanted areas
kernel = np.ones((3,3), np.uint8)
v_chan = cv2.erode(v_chan, kernel, iterations=2)
v_chan = cv2.dilate(v_chan, kernel, iterations=4)

# Perform a connected component analysis
nlabels, labels, stats, _ = cv2.connectedComponentsWithStats(v_chan)

# Create a mask to store the points of interest
poi_mask = np.zeros(v_chan.shape, dtype="uint8")

# Create a variable to store the number of points
poi = 0

# Loop over the labels
for label in np.unique(labels):
    # Ignore the background label
    if label == 0:
        continue

    # Build the label mask and count the number of pixels in the area
    label_mask = np.zeros(v_chan.shape, dtype="uint8")
    label_mask[labels == label] = 255
    num_pixels = cv2.countNonZero(label_mask)

    # if the number of pixels in the component are sufficient in number then add the area to the point of interest mask
    if num_pixels > 50:
        poi_mask = cv2.add(poi_mask, label_mask)
        poi += 1

print ('Points of interest:', poi)
# Calculate the percentage of correctly detected red regions
total_red_pixels = np.count_nonzero((red_mask_final > 0).astype(np.uint8))
correctly_detected_red_pixels = np.count_nonzero(np.logical_and(red_mask_final > 0, poi_mask > 0).astype(np.uint8))
accuracy_percentage = (correctly_detected_red_pixels / total_red_pixels) * 100

print('Accuracy Percentage:', accuracy_percentage, '%')

#Convert the POI mask back to RGB before overlaying the source image
poi_mask_back_to_rgb = cv2.cvtColor(poi_mask, cv2.COLOR_GRAY2RGB)
overlay = cv2.addWeighted(source, 0.4, poi_mask_back_to_rgb, 0.6, 0)

cv2.imshow('Source Image', source)
cv2.imshow('Points of Interest Mask', poi_mask)
cv2.imshow('Overlayed Image', overlay)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Red Object Detection and Highlighting

This code aims to detect and highlight red objects within an image using the OpenCV library. Here's a breakdown of the code:

1. **Import Necessary Libraries:**
   - `cv2`: OpenCV library for image processing.
   - `numpy` (aliased as `np`): Used for numerical operations on arrays.
  
2. **Import the Image:**
   - It loads an image named 'project.jpg' using the `cv2.imread()` function. You should replace 'project.jpg' with the path to your own image.

3. **Remove Noise:**
   - To enhance the image, a median blur with a kernel size of 15x15 is applied using `cv2.medianBlur()`.

4. **Convert to HSV Color Space:**
   - The image is converted from the standard BGR (Blue-Green-Red) color space to the HSV (Hue-Saturation-Value) color space using `cv2.cvtColor()`. HSV is useful for color-based image processing.

5. **Define Red Masks:**
   - Two masks are defined to detect red color:
     - `red_mask_1` for the lower range of red hues (0-10 in HSV).
     - `red_mask_2` for the upper range of red hues (160-180 in HSV).
     These masks are created using `cv2.inRange()`.

6. **Combine Masks:**
   - The two red masks are combined using element-wise addition to create `red_mask_final`.

7. **Bitwise AND with the HSV Image:**
   - A bitwise AND operation is performed between `hsv_source` (HSV image) and `red_mask_final` to obtain `output_hsv`.
 it compares the corresponding pixels of two images and produces a new image where each pixel is 1 if both pixels in the original images are 1, otherwise 0. It's often used in image processing to create masks and perform binary operations
8. **Blur the Image:**
   - A Gaussian blur is applied to `output_hsv` to smoothen out bright areas using `cv2.GaussianBlur()`.

9. **Thresholding:**
   - A thresholding operation is performed to reveal bright areas by converting the blurred image into a binary mask. Pixels with intensity values greater than 60 are set to 255 (white), and the rest to 0 (black).

10. **Split Image Components:**
    - The thresholded image is split into its hue (`h_chan`), saturation (`s_chan`), and value (`v_chan`) components using `cv2.split()`.

11. **Erode and Dilate:**
    - The value component (`v_chan`) undergoes morphological operations to remove small unwanted areas. Erosion followed by dilation is performed using a 3x3 kernel.

12. **Connected Component Analysis:**
    - The connected components in the processed image (`v_chan`) are identified using `cv2.connectedComponentsWithStats()`. This function returns information about each connected component.

13. **Create a Mask for Points of Interest:**
    - A binary mask named `poi_mask` is created to store points of interest (potential red objects).

14. **Count and Filter Components:**
    - The code counts and filters connected components based on the number of pixels in each component. Components with fewer than 50 pixels are ignored, while larger components are added to `poi_mask`.

15. **Display Results:**
    - The code prints the number of detected points of interest (red objects).
    - The `poi_mask` is converted back to RGB format (`poi_mask_back_to_rgb`) for overlaying.
    - The original image, the points of interest mask, and the overlay of the original image and points of interest are displayed using `cv2.imshow()`. The images are displayed until a key is pressed, and then the windows are closed using `cv2.destroyAllWindows()`.

This code can be adapted to detect and highlight red objects in an image while excluding orange ones by adjusting the HSV range and pixel threshold for component filtering.
